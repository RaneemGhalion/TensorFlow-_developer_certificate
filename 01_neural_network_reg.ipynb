### how mgiht we build a model ?

###![VS--CourseTensorFlowDeveloperCertificateBootcampUdemy-4’39”.jpg](VS--CourseTensorFlowDeveloperCertificateBootcampUdemy-10’54”.jpg) steps in modeling with tensorflow

```python
# create a model - define the input and output layers ,as well as the hidden layers of a deep learning model
# compiling the midel - define the loss function and the optimizer(tells our model how to improve the patterns its learning) and evaluation metrics
# fitting the model - letting the model trying to find the patterns between X and Y
```

```python
import tensorflow as tf
tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()

# Assuming X and Y are 1D arrays (for example: X = [a, b, c,...], Y = [d, e, f,...])
X = tf.constant(X, dtype=tf.float32)  # Convert to tensor if they're not already
Y = tf.constant(Y, dtype=tf.float32)


# Create the model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])  # Assuming each input sample has 1 feature
])

# Compile the model
model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.SGD(), metrics=["mae"])

# Fit the model
model.fit(X, Y, epochs=100)

# Try and make a prediction using our model
# Ensure the input to predict is in the same shape as X
# model.predict([17.0])
model.predict(tf.reshape([17.0], (1, 1)))

```

```python
import tensorflow as tf
tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()

# Assuming X and Y are 1D arrays (for example: X = [a, b, c,...], Y = [d, e, f,...])
X = tf.constant(X, dtype=tf.float32)  # Convert to tensor if they're not already
Y = tf.constant(Y, dtype=tf.float32)


# Create the model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation=None,input_shape=[1]),
    tf.keras.layers.Dense(1, input_shape=[1])  # Assuming each input sample has 1 feature
])

# Compile the model
model.compile(loss=tf.keras.losses.mae, optimizer=tf.keras.optimizers.Adam(learning_rate=0.01), metrics=["mae"])
# Fit the model
model.fit(X, Y, epochs=100)

# Try and make a prediction using our model
# Ensure the input to predict is in the same shape as X
# model.predict([17.0])
model.predict(tf.reshape([17.0], (1, 1)))

```
