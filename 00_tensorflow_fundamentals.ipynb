# Indexing tensors dot_product.jpg

```python
# Get the first 2 elements of each dimension of each diminsion of tensor

rank_4_tensor=tf.zeros(shape=[2,3,4,5])
```

```python
rank_4_tensor[:2,:2,:2,:2]
```

```python
# Get the first each dimension from each index except final one

rank_4_tensor[:1,:1,:1,:]
```

```python
rank_4_tensor[:1,:,:1,:1]
```

```python
# adding extra dimensions
rank_2_tensor=tf.constant([[10,7],[3,4]])
rank_2_tensor.shape ,rank_2_tensor.ndim
```

```python
# Get the last item of each row of our rank 2 tensor

rank_2_tensor
```

```python
rank_2_tensor[1,:]
```

```python
# adding extra dimensions
rank_3_tensor=rank_2_tensor[...,tf.newaxis,tf.newaxis]
rank_3_tensor
```

```python
# alternative to tf.newaxis
tf.expand_dims(rank_2_tensor,axis=-1)

# Expand dimensions to add an extra dimension
expanded_tensor = tf.expand_dims(rank_2_tensor, axis=-1) # Now shape is (2, 2, 1)
print(expanded_tensor)
print("----------------")
# Duplicate the data along the new dimension to get shape (2, 2, 2)
result_tensor = tf.tile(expanded_tensor, [1, 1, 2]) # Now shape is (2, 2, 2)

print(result_tensor)
```

```python
print(result_tensor.shape)
```
