# random tensors and shuffle the elements in a tensor

# why? order can affect nueral netwrok learning






```python
# shuffle a tensor (shuffle your data so the inherent order not affect learning)

not_shuffled =tf.constant([[10,7],[3,4],[2,5]])

not_shuffled
```

```python
not_shuffled.ndim
```

```python
# let's shuffle
tf.random.set_seed(42)
tf.random.shuffle(value=not_shuffled,seed=42)  #shuffle on its first dimension
```
